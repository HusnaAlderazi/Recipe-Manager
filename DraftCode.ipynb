```python
import csv
# 1. VIEW ALL RECIPES
def view_recipes():
    file = open("recipes.csv", "r", encoding="utf-8")
    reader = csv.DictReader(file)

    print("\n--- List Of Recipes ---")
    i = 1
    for row in reader:
        print(str(i) + ". " + row["name"] + " - " + row["preparation_time"] + " minutes")
        i += 1

    print()
    file.close()
```
